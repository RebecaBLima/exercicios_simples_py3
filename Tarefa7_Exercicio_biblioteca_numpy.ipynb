Escreva um programa em Python utilizando a biblioteca Numpy para resolver o sistema de equações abaixo. 


```python
#aqui impoetamos a biblioteca utilizada
import numpy as np
```

```python
#Resolvendo o sistema de equações acima

a = np.array([[1,1,1,-1], [2,1,-2,1], [1,1,1,1], [1,-2,1,2]])
b = np.array([6,-1,2,-3])

x = np.linalg.solve(a, b)

print('A solução pela função linalg.solve é:', x)
print('Verificando se esta solução acima é correta:', np.allclose(np.dot(a, x), b))
```
