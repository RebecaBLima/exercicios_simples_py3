A Figura abaixo ilustra o funcionamento de um circuito composto por uma fonte de tensão, um resistor de resistência R e um indutor de indutância L.



Sabendo-se que a tensão é dada por V(t) = V0 sen(ωt), com V0 = 50V e ω = 120π Hz, a resistência R = 300Ω e a indutância L = 1H.

Utilize a biblioteca SymPy para determinar a corrente elétrica em função do tempo e ilustrá-la em um gráfico.

```python
%matplotlib inline 
```

```python
import sympy
import numpy as np
from sympy import*
i = Function("i")
t = Symbol('t')
i_ = Derivative(i(t), t)
L=Symbol('L')
R=Symbol('R')
V=Symbol('V')
V0=Symbol('V0')
omega=Symbol('𝜔')
Eq(L*i_+R*i(t)-V)
```

```python
Eq(L*i_+R*i(t)-V0*sympy.sin(omega*t))
```

```python
dsolve(L*i_ + R*i(t) - V, i(t))
```

```python
dsolve(Eq(L*i_ + R*i(t) - V), 0)
```

```python
a = dsolve(L*i_+R*i(t)-V0*sympy.sin(omega*t))
a
```

```python
b = Eq(L*i_+R*i(t)-V0*sympy.sin(omega*t))
```

```python
i0 = a.subs({t:0, V:0, R:300, L:1, i:0, omega:120*np.pi, V0:50})
i0
```

```python
b1=dsolve(b, i(t))
b1 
```
