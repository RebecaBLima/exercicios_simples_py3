## Gravitação Universal de Newton

Antes de introduzir as Leis de Kepler, é importante explicar um pouco sobre a Lei da Gravitação Universal Newtoniana .... continuar aqui

O seu peso, a força que te atrai para o centro da Terra, talvez seja o mais familiar exemplo de atraçãogravitacional que você conhece. Estudando o movimento da Lua e dos planetas, analisando as teorria de movimentos anteriores a sua época (como as Leis de Kepler - movimento da Lua e planetas), Newton descobriu o caráter fundamental da atração gravitacional entre dois corpos de qualquer natureza. Juntamente com as 3 leis de movimento, Newton publicou a lei da gravitação em 1687. Ela pode ser enunciada do seguinte modo: 

***Cada partícula do universo atrai qualquer outra partícula com uma força diretamente proporcional ao produto de suas respectivas massas e inversamente proporcional ao quadrado da distância entre as partículas.***

Matematicamente, isso quer dizer que 
\begin{equation}
F_g=\left (\frac{GMm}{r^2}\right)
\end{equation}
onde $F_g$ é o módulo da força que atua sobre cada partícula, $M$ e $m$ são as massas das partículas, e $r$ é a distância entre elas e $G$ é uma constante fundamental denominada **constante gravitacional.**

## Leis de Kepler

Em 1609, Johannes Kepler publicou _Astronomia Nova… De Motibus Stellae Martis,_ onde apresentou as três leis do movimento dos planetas, que hoje levam seu nome:

* Os planetas descrevem órbitas elípticas, com o sol num dos focos;
* O raio vetor que liga um planeta ao Sol descreve áreas iguais em tempos iguais. (lei das áreas);
* Os quadrados dos períodos de revolução $T$ são proporcionais aos cubos das distâncias médias $a$ do Sol aos planetas. $T^{2}=ka^{3}$, onde $k$ é uma constante de proporcionalidade.

#### Dedução das leis de Kepler
Com o Princípio da Gravitação Universal elaborado por Isaac Newton, foi possível postular um único princípio:

que, aliado às Três Leis de Newton, foi capaz de explicar completamente as observações astronômicas conhecidas até a época e ainda depois, até a descoberta de que a velocidade da luz no vácuo é constante para todos os referenciais. Essa descoberta levou à criação da teoria da relatividade restrita e, consequentemente, da teoria da relatividade geral, que, para certos fenômenos que até então não haviam sido observados, invalida a teoria de Newton da gravitação.

No entanto, as leis de Newton e a sua teoria da gravidade são mais do que o suficientes para explicar as leis de Kepler. De fato, as três leis são deriváveis da simples equação postulada acima, de modo que ainda aparecem mais completas do que da forma descrita por Kepler.

Para derivá-las, é preciso introduzir algunsconceitos.

$\dot  x$ representa a derivada temporal de x, enquanto $\ddot  x$ é a derivada temporal segunda de $x$.
$\hat  {r}$ é o vetor unitário que indica a direção do planeta em relação à sua estrela. A derivada temporal desse vetor, que representaremos como $\dot{\hat{r}}$ é igual a ${\dot {\theta }}.{\hat {\theta }}$, onde ${\dot  {\theta }}$ é a velocidade angular do planeta em relação à estrela, e ${\hat  {\theta }}$ é um vetor unitário perpendicular a ${\hat  {r}}$. Existem duas direções possíveis de um vetor unitário perpendicular a outro, mas a direção deste é escolhida de modo que ${\hat  {r}}$ tivesse que virar no sentido anti-horário para apontar na mesma direção dele. A derivada de ${\hat  {\theta }}$, por sua vez, é $-{\dot {\theta }}.{\hat {r}}$.

O vetor ${\vec  r}$ é o vetor-posição do planeta em relação à sua estrela, e é definido como ${\vec  r}=r{\hat  {r}}$, onde $r$ é o módulo da distância entre o planeta e a estrela. Assim, ${\displaystyle {\dot {\vec {r}}}={\dot {r}}{\hat {r}}+r{\dot {\hat {r}}}}{\dot {{\vec r}}}={\dot r}{\hat {r}}+r{\dot {{\hat {r}}}}$. Seguindo daí,

\begin{equation}
\ {\ddot {{\vec r}}}={\ddot r}{\hat {r}}+{\dot r}{\dot {{\hat r}}}+{\dot r}{\dot {\theta }}{\hat {\theta }}+r{\ddot {\theta }}{\hat {\theta }}+r{\dot {\theta }}{\dot {{\hat {\theta }}}} {\displaystyle {\ddot {\vec {r}}}={\ddot {r}}{\hat {r}}+{\dot {r}}{\dot {\theta }}{\hat {\theta }}+{\dot {r}}{\dot {\theta }}{\hat {\theta }}+r{\ddot {\theta }}{\hat {\theta }}-r{\dot {\theta }}{\dot {\theta }}{\hat {r}}}{\ddot {{\vec r}}}={\ddot r}{\hat {r}}+{\dot r}{\dot {\theta }}{\hat {\theta }}+{\dot r}{\dot {\theta }}{\hat {\theta }}+r{\ddot {\theta }}{\hat {\theta }}-r{\dot {\theta }}{\dot {\theta }}{\hat {r}}.
\end{equation}
 
Organizando os termos, temos, 

\begin{equation}
{\displaystyle {\ddot {\vec {r}}}=({\ddot {r}}-r{\dot {\theta }}^{2}){\hat {r}}+(r{\ddot {\theta }}+2{\dot {r}}{\dot {\theta }}){\hat {\theta }}}{\ddot {{\vec r}}}=({\ddot r}-r{\dot \theta }^{2}){\hat r}+(r{\ddot \theta }+2{\dot r}{\dot \theta }){\hat {\theta }}\
\end{equation}

Essas deduções serão utilizadas nas leis apresentadas a seguir:

### Primeira Lei de Kepler
" *O planeta em órbita em torno do Sol descreve uma elipse em que o Sol ocupa um dos focos.* "

Esta lei definiu que as órbitas não eram circunferências, como se supunha até então, mas sim elipses.

A primeira lei de Kepler é uma observação empírica de que os planetas se movem em órbitas elípticas. Newton provou mais tarde, usando sua forma de cálculo geométrico, que em todas as órbitas elípticas, a força que atrai o corpo orbital para os focos da elipse é o quadrado inverso da distância. Newton também mostrou que o inverso disso é verdadeiro, que corpos orbitais governados por uma força do inverso do quadrado seguem um caminho elíptico. Neste projeto você irá ver a demonstração que dado um corpo em movimento e uma força inversa do quadrado em direção a um objeto central, o corpo pode seguir um caminho elíptico.


Considerando um planeta como sendo uma partícula (o que se justifica com boa aproximação para o fim das leis de kepler, já que o tamanho dos planetas do sistema solar são desprezíveis em comparação com a sua distância ao sol). Então, usamos a Lei da Gravitação Universal 
\begin{equation}
{\vec F}=-G{\frac {Mm}{r^{2}}}{\hat {r}}
\end{equation}

Supondo que a massa do planeta é constante, (o que está de acordo com os sistemas observados por kepler), usamos a Segunda Lei de Newton.



\begin{equation}
m{\ddot {{\vec r}}}=-G{\frac {Mm}{r^{2}}}{\hat {r}} \\
{\ddot {{\vec r}}}=-G{\frac {M}{r^{2}}}{\hat {r}} \\
({\ddot {r}}-r{\dot {\theta }}^{2}){\hat {r}}+(r{\ddot {\theta }}+2{\dot {r}}{\dot {\theta }}){\hat {\theta }}=G{\frac {M}{r^{2}}}{\hat {r}}
\end{equation}

Assim,

\begin{equation}
{\ddot r}-r{\dot \theta }^{2}=-G{\frac {M}{r^{2}}}
\end{equation}

e

\begin{equation}
r{\ddot \theta }+2{\dot r}{\dot \theta }=0
\end{equation}


Da última, podemos derivar a conservação do momento angular, multiplicando os dois membros por $mr$:

\begin{equation}
mr^{2}{\ddot  \theta }+2mr{\dot  r}{\dot \theta }=0 \\
{\frac  {d}{dt}}[mr^{2}{ \dot {\theta }}]=0 \\
mr^{2}{\dot  {\theta }}=l
\end{equation}

onde $l$ é uma constante, que sabemos ser a magnitude do momento angular.

Podemos transformar derivadas temporais em derivadas em relação a $\theta$ , a partir da seguinte relação:

\begin{equation}
{\frac  {d\theta }{dt}}={\frac  {l}{mr^{2}}}
\end{equation}
Se tivermos a derivada de qualquer função $X(t)$ em relação ao tempo, podemos usar a regra da cadeia:

\begin{equation}
{\frac  {dX}{dt}}={\frac  {dX}{d\theta }}{\frac  {d\theta }{dt}} \\
{\frac  {dX}{dt}}={\frac  {l}{mr^{2}}}{\frac  {dX}{d\theta }}
\end{equation}
O que é de grande utilidade na equação diferencial.


Para a primeira equação temos:

\begin{equation}
{\ddot  r}-r{\dot  \theta }^{2}=-G{\frac  {M}{r^{2}}} \\
{\frac  {l}{mr^{2}}}{\frac  {d}{d\theta }}\left[{\frac  {l}{mr^{2}}}{\frac  {dr}{d\theta }}\right]-r{\dot  \theta }^{2}=-G{\frac  {M}{r^{2}}} \\
\left(-2{\frac  {l}{mr^{3}}}\left({\frac  {dr}{d\theta }}\right)^{2}+{\frac  {l}{mr^{2}}}{\frac  {d^{2}r}{d\theta ^{2}}}\right)-r{\dot  \theta }^{2}=-G{\frac  {M}{r^{2}}} \\
-2{\frac  {l^{2}}{m^{2}r^{5}}}\left({\frac  {dr}{d\theta }}\right)^{2}+{\frac  {l^{2}}{m^{2}r^{4}}}{\frac  {d^{2}r}{d\theta ^{2}}}-r{\dot  \theta }^{2}=-G{\frac  {M}{r^{2}}}
\end{equation}

É preciso aqui extrair do momento angular uma relação útil:
\begin{equation}
{\displaystyle l=mr^{2}{\dot {\theta }}} \\
{\displaystyle {\dot {\theta }}={\frac {l}{mr^{2}}}}
\end{equation}

Substituindo na equação principal,

\begin{equation}
\left(-2{\frac {l}{mr^{3}}}\left({\frac {dr}{d\theta }}\right)^{2}+{\frac {l}{mr^{2}}}{\frac {d^{2}r}{d\theta ^{2}}}\right)-r{\dot \theta }^{2}=-G{\frac {M}{r^{2}}}
\end{equation}

Aqui, convém usar uma transformação de variável:

\begin{equation}
 u=r^{-1} \\
 {\displaystyle {\frac {du}{d\theta }}=-r^{-2}{\frac {dr}{d\theta }}} \\
 {\displaystyle {\frac {d^{2}u}{d\theta ^{2}}}=2r^{-3}\left({\frac {dr}{d\theta }}\right)^{2}-r^{-2}{\frac {d^{2}r}{d\theta ^{2}}}}
\end{equation}
Utilizando-a na equação diferencial, a simplificamos significativamente.

\begin{equation}
-{\frac  {l^{2}}{m^{2}r^{2}}}{\frac  {d^{2}u}{d\theta ^{2}}}-{\frac  {l^{2}}{m^{2}r^{3}}}=-G{\frac  {M}{r^{2}}} \\
{\displaystyle -{\frac {l^{2}}{m^{2}}}{\frac {d^{2}u}{d\theta ^{2}}}-{\frac {l^{2}}{m^{2}r}}=-GM} \\
{\frac  {d^{2}u}{d\theta ^{2}}}+u={\frac  {GMm^{2}}{l^{2}}}
\end{equation}
A função que satisfaz à essa equação diferencial é:

\begin{equation}
r={\frac  {l^{2}}{GMm^{2}}}{\frac  {1}{1+\epsilon \cos(\theta +{\theta }_{{0}})}}
\end{equation}

$\epsilon$  é uma constante arbitrária de integração, e pode ser obtido se for dada a posição do planeta em qualquer instante. Com $\epsilon$  menor do que $1$, temos a equação de uma elipse escrita em coordenadas polares. Se $\epsilon$ for  $0$, a equação é a de um círculo.
Assim, derivamos a Primeira Lei de Kepler.

A seguir, você verá simulações de órbitas planetárias que obedecem a Primeira Lei de Kepler.

```python
#aqui possibilita que os gráficos sejam mostrados nesse ambiente
%matplotlib inline 

#aqui importamos as bibliotecas que serão utilizadas
import numpy as np 
import matplotlib.pyplot as plt
```

Agora vamos definir algumas variáveis iniciais para determinar como esta simulação se comportará. A `altura` é o quão acima do centro o corpo começará. `velocidade` é o quão rápido o corpo estará indo horizontalmente. O `step` é o número de atualizações pelas quais a simulação ocorrerá. `c` armazena as coordenadas do centro da órbita. `corpo` armazena as coordenadas da posição inicial do corpo. E `v` é a velocidade inicial do corpo.

```python
#aqui definimos os parâmetros o gráfico das órbitas
altura = -254.0 - 3959
velocidade = 4.76
steps = 10000 #aqui podemos definir o números de órbitas no gráfico, quanto menor o número, menor será a quantidade de órbitas
c = (0.0,0.0)
corpo = (0.0, altura)
v = (velocidade, 0.0)
```

```python
# aqui denfinimos x e y
x = []; y = [];
```

Agora é hora de começar a simular as órbitas. A ideia geral é que as forças no corpo serão calculadas e então o corpo será movido de acordo com essas forças. Isso acontecerá em n etapas. Após a conclusão de cada etapa, a posição resultante do corpo é armazenada nas listas x e y.

```python
#Aqui adicionamos as posições iniciais para às matrizes
x.append(corpo[0])
y.append(corpo[1])

#Aqui leva n passos 
for i in range(1,steps):
    
    #Aqui encontramos a distância que o corpo está do centro usando o Teorema de Pitágoras 
    dx = c[0]-corpo[0]
    dy = c[1]-corpo[1]
    d = np.sqrt(dx*dx + dy*dy)
    
    #Aqui calculamos a força que atua no corpo usando a lei do inverso do quadrado 
    f = 1/d*d
    
    #Aqui encontramos o vetor de força em direção ao centro da órbita
    direcao = (dx, dy)
    
    #Aqui obtemos o comprimento deste vetor para que possamos torná-lo de comprimento unitário
    comprimento_direcao = np.sqrt(direcao[0]**2 + direcao[1]**2)
    
    #Aqui temos a direção normal
    direcao = (dx/comprimento_direcao, dy/comprimento_direcao)
    
    #Aqui podemos multiplicar o vetor de direção de comprimento unitário pelo comprimento da força para obter o vetor de força # 
    vetor_forca = (f*direcao[0],f*direcao[1])
    
    #Redefinindo a velocidade do corpo
    v = (v[0]+vetor_forca[0],v[1]+vetor_forca[1])
    
    #Redefinindo a posição do corpo
    corpo = (corpo[0]+v[0],corpo[1]+v[1])

    #Gravando a posição
    x.append(corpo[0])
    y.append(corpo[1])
```

Agora que a simulação foi realizada, é hora de plotar os dados. No entanto, precisamos primeiro converter nossas listas em matrizes  da biblioteca numpy.

```python
x = np.array(x)
y = np.array(y)
plt.plot(x,y)
plt.title('Gráfico 1 - Resultados da Simulação das Órbitas');
```

Como pode ser visto no gráfico resultante, as órbitas parecem seguir um caminho elíptico. Isso significa que, usando a linguagem Python, podemos verificar as descobertas de Kepler e Newton. Para obter um padrão de órbita diferente, basta alterar algumas das variáveis iniciais.

###### Exemplo 2: Órbita Terrestre
Agora, vamos simular a órbita terreste. Para fazer isso, precisamos especificar todas as variáveis usadas na função `Orbita_de_Kepler`. Tomando nota das unidades, vamos inserir o período orbital e excentricidade da Terra.

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
#aqui definimos a função da órbita com os focos
def Terceira_Lei_de_Kepler_eixo_maior(periodo_orbital):
   

    #Unidades: período orbital [unidade em ano], distância [unidade em UA]           
    
    cubo_do_semieixo_maior = periodo_orbital**2
    semieixo_maior = cubo_do_semieixo_maior**(1./3.)
    
    return semieixo_maior

def Orbita_de_Kepler(excentricidade,periodo_orbital):

    #Unidades: Período orbital em anos                          
   
    nStep = 500
    tRange = np.linspace(0.0,periodo_orbital,nStep)
    
    theta = []    
    for tempo in tRange:
        PsiDeriv = 1.0
        M = 2*np.pi*tempo/periodo_orbital
        Psii = M
        theta0i = 180.0
        while PsiDeriv > 1e-10:
            Psii= M + excentricidade*np.sin(Psii)
            Psif = Psii
            PsiDeriv = Psif-Psii
            
        theta0 = 2*np.arctan(((1+excentricidade)/(1-excentricidade))**(0.5)*np.tan(Psii/2.))
        theta.append(theta0)        
    return theta
```

```python
periodo_orbital = 1.0 #unidade em anos
excentricidade = 0.02 #valores arredondados
```

Agora estamos prontos para fazer nossa órbita de Kepler. Para fazer isso, precisamos enviar as variáveis certas para a função e estar prontos para a variável que a função envia de volta. `Orbita_de_Kepler` envia de volta uma lista de ângulos; esses ângulos são diferentes ângulos do `angulo` escolhido para cada vez na órbita.

```python
list1 = Orbita_de_Kepler(excentricidade,periodo_orbital)
```

Para fazer gráficos, precisamos ser capazes de especificar a posição x e y de cada valor do `angulo`. Isso é feito usando a equação da forma de kepler: 
\begin{equation}
 r=\left (\frac{a(1-e^2)}{1+ecos(\theta)}\right)
\end{equation}


```python
def orbita(semi_eixo_maior,excentricidade,angulo): 

    #Unidades: em [UA] 
    
    #Definindo a equação de forma de Kepler
    r_orbita = semi_eixo_maior*(1 - excentricidade**2)/(1 + excentricidade*np.cos(angulo))
    x_orbita = r_orbita*np.cos(angulo)
    y_orbita = r_orbita*np.sin(angulo)
     
    
    return x_orbita,y_orbita
```

Na equação de forma, theta é o `angulo`.
A função abaixo, orbita (`semi_eixo_maior`, e, `angulo`), calcula os valores x e y da equação de forma e os retorna como `xorbita` e `yorbita`. Podemos usar esses valores x e y para fazer um gráfico de dispersão da órbita.
Observe que a função de órbita precisa do semi-eixo principal. Isso significa que precisaremos usar a função da terceira lei de Kepler para converter o período orbital em separação.

```python
semi_eixo_maior = Terceira_Lei_de_Kepler_eixo_maior(periodo_orbital)
```

Agora que temos a separação, podemos chamar a função de órbita. A função de órbita precisa da separação, `excentricidade` e `angulo` para funcionar. Lembre-se de que o angulo é retornado pela função `Orbita_de_Kepler.`

```python
xOrbita, yOrbita = orbita(semi_eixo_maior,excentricidade,list1)
```

### Segunda Lei de Kepler

“ *A linha que liga o planeta ao Sol varre áreas iguais em tempos iguais.* ”

Esta lei determina que os planetas se movem com velocidades diferentes, dependendo da distância a que estão do Sol.
Periélio é o ponto mais próximo do Sol, onde o planeta orbita mais rapidamente.
Afélio é o ponto mais afastado do Sol, onde o planeta move-se mais lentamente.

A segunda Lei de Kepler é bem mais simples de se derivar.

A área descrita pelo raio-vetor que liga o planeta à sua estrela durante um certo tempo é dada por
\begin{equation}
\sum _{{i}}^{{N}}a_{i}
\end{equation}

Onde $a_i$ são as áreas percorridas em frações desse tempo. Podemos fazer essas frações de tempo arbitrariamente pequenas, e consequentemente teremos um $N$ cada vez maior. Nada se altera se fizermos o limite em que as frações de tempo tendem a $0$, ou seja:
\begin{equation}
A=\lim _{{N\to \infty }}\sum _{{i}}^{{N}}a_{i}.
\end{equation}

Quando tomamos áreas $a_i$ menores, elas se aproximam arbitrariamente da área de um triângulo com base $\Delta r_{i}$ e altura $r_i$ , onde $r_i$ é a magnitude do raio vetor ${\vec  r}_{i}$ que liga o planeta à sua estrela em algum instante dentro de um intervalo de tempo $[t,t+\Delta t]$, e $\Delta r_{i}=|{\vec  r}_{i}-{\vec  r}_{{i-1}}|$, com ${\vec  r}_{{i-1}}$ sendo o análogo de $r_i$ em algum instante dentro do intervalo $[t-\Delta t,t]$. Ou seja, $\Delta r_{i}$ é simplesmente a distância percorrida pelo planeta em um certo tempo.

Ou seja, as áreas $a_i$ se aproximam arbitrariamente de:

${\frac  {r_{i}.\Delta r_{i}}{2}}$
$\Delta r_{i}$ também pode se expressado como $\Delta t$, onde $v_{i}$ é a velocidade do planeta, em algum instante do mesmo intervalo de tempo de $r_i$.
Quando $N$ tende a infinito,$\Delta t$ tende a $0$. Assim,
\begin{equation}
\lim _{{N\to \infty }}\sum _{{i}}^{{N}}a_{i}=\lim _{{N\to \infty }}\sum _{{i}}^{{N}}{\frac  {r_{i}.v_{i}\Delta t}{2}}
\end{equation}
O que constitui uma integral:
\begin{equation}
A=\int _{{0}}^{{t}}{\frac  {rv}{2}}dt
\end{equation}

ou, como $v=r{\dot  {\theta }}$,
\begin{equation}
{\displaystyle A=\int _{0}^{t}{\frac {r^{2}{\dot {\theta }}}{2}}dt}
\end{equation}

$r^{2}{\dot  {\theta }}$ é o momento angular sobre a massa, o que nesse caso permanece sempre constante. Assim, a integral dá:
\begin{equation}
A={\frac  {l}{2m}}t
\end{equation}
Como o momento angular é sempre o mesmo, são percorridas áreas iguais em tempos iguais. Temos a segunda Lei de Kepler.

### Terceira Lei de Kepler


“ *Os quadrados dos períodos de translação dos planetas são proporcionais aos cubos dos semi-eixos maiores de suas órbitas.*	”
Ou seja, sendo $T$ o período de revolução (ano do planeta) e $D$ o semi-eixo maior da órbita de um planeta, tem-se:
\begin{equation}
\frac {D^3}{T^2}=k 
\end{equation}
com $k$ sendo uma constante.
Esta lei indica que existe uma relação entre a distância do planeta e o período de translação (tempo que ele demora para completar uma revolução em torno do Sol). Portanto, quanto mais distante estiver do Sol mais tempo levará para completar sua volta em torno desta estrela.



A terceira Lei de Kepler é mais sutil. Ela é escrita em função do raio médio, então devemos achar esse raio. Na equação do raio:
\begin{equation}
r={\frac  {l^{2}}{GMm^{2}}}{\frac  {1}{1+\epsilon \cos(\theta +{\theta }_{{0}})}}
\end{equation}
A única variável é o $\cos(\theta +\theta _{0})$, de modo que o raio médio corresponde ao valor médio dessa variável. Esse valor corresponde a
$\langle \cos(\theta +\theta _{0})\rangle ={\frac  {1}{2\pi }}\int _{{0}}^{{2\pi }}\cos(\theta +{\theta }_{{0}})d\theta ={\frac  {1}{2\pi }}(\sin(2\pi +\theta )-\sin(\theta ))=0$ . Assim, o raio médio correspondente é:
\begin{equation}
{r}_{{med}}={\frac  {l^{2}}{GMm^{2}}}
\end{equation}

Podemos pensar também na velocidade angular média, correspondente ao raio médio. Ambos estão ligados através do momento angular $({\displaystyle l=m{\dot {\theta }}_{med}{r}_{med}^{2}}l=m{\dot  {\theta }}_{{med}}{r}_{{med}}^{2})$. Então
\begin{equation}
l=m{{\dot  {\theta }}_{{med}}}{\frac  {l^{4}}{G^{2}M^{2}m^{4}}} \\
{{\dot  {\theta }}_{{med}}}={\frac  {G^{2}M^{2}m^{3}}{l^{3}}}
\end{equation}
Uma definição importante é a do período, em função da velocidade angular média:

\begin{equation}
P={\frac  {2\pi }{{{\dot  {\theta }}}_{{med}}}}
\end{equation}

A presença da velocidade angular média nessa equação é justificada pelo fato de que deve haver algum valor da velocidade angular, em algum instante, que satisfaça a essa equação. Esse valor é justamente o da velocidade angular média.

É possível demonstrar que o período de um planeta com órbita circular de raio ${r}_{med}$ e velocidade angular ${{\dot  {\theta }}_{{med}}}$ é igual ao período de qualquer planeta com órbita elíptica de raio médio ${r}_{{med}}$ e velocidade angular média ${{\dot  {\theta }}_{{med}}}$. Isso é feito através da Segunda Lei de Kepler. A área total de um círculo é $\pi {r}_{{med}}^{2}$, e a área total de uma elipse é, pela Segunda Lei, ${\frac  {l}{2m}}P$. Através da definição de $P$ acima, vemos:

\begin{equation}
A={\frac  {l}{2m}}P \\
A={\frac  {l\pi }{{m{\dot  {\theta }}}_{{med}}}} \\
A={\frac  {l^{4}\pi }{G^{2}M^{2}m^{4}}}
\end{equation}

Lembrando a equação correspondente ao raio médio ${r}_{med}=\frac{l^2}{GMm^2}$, temos:
\begin{equation}
A={r}_{{med}}^{2}\pi
\end{equation}
Que corresponde à área do círculo. Como, pela Segunda Lei, áreas iguais são percorridas em tempos iguais, então o período do planeta de órbita elíptica pode ser tomado a partir do período de uma órbita circular correspondente.

\begin{equation}
P={\frac  {2\pi }{{{\dot  {\theta }}_{{med}}}}} \\
P={\frac  {2\pi l^{3}}{G^{2}M^{2}m^{3}}} \\
P={\frac  {2\pi {r}_{{med}}^{{{\frac  {3}{2}}}}}{{\sqrt  {GM}}}} \\
P^{2}={\frac  {4\pi ^{2}{r}_{{med}}^{3}}{GM}} \\
{\frac  {P^{2}}{{r}_{{med}}^{3}}}={\frac  {4\pi ^{2}}{GM}} \\
\end{equation}
O que constitui a terceira lei de Kepler.



## Problema de Dois Corpos

Em mecânica clássica, o **problema dos dois corpos** é prever o movimento de dois objetos massivos que são vistos abstratamente como partículas pontuais. O problema assume que os dois objetos interagem apenas um com o outro; a única força que afeta cada objeto surge do outro, e todos os outros objetos são ignorados.

O caso mais proeminente do problema clássico de dois corpos é o caso gravitacional, surgindo na astronomia para prever as órbitas (ou fugas da órbita) de objetos como satélites, planetas e estrelas. Um modelo de partícula de dois pontos desse sistema quase sempre descreve seu comportamento bem o suficiente para fornecer insights e previsões úteis.

Um modelo mais simples de "um corpo", o "problema da força central", trata um objeto como a fonte imóvel de uma força agindo sobre outro. Em seguida, procura-se prever o movimento do único objeto móvel remanescente. Tal aproximação pode fornecer resultados úteis quando um objeto é muito mais massivo do que o outro (como com um planeta leve orbitando uma estrela pesada, onde a estrela pode ser tratada como essencialmente estacionária).

No entanto, a aproximação de um corpo geralmente é desnecessária. Para muitas forças, incluindo as gravitacionais, a versão geral do problema de dois corpos pode ser reduzida a um par de problemas de um corpo, permitindo que sejam resolvidos completamente e fornecendo uma solução simples o suficiente para ser usada com eficácia.

Em contraste, o problema de três corpos (e, mais geralmente, o problema de n-corpos para n ≥ 3) não pode ser resolvido em termos de primeiras integrais, exceto em casos especiais.



### Enunciado do Problema

Sejam A e B corpos de tamanho desprezível, de massas, respectivamente, M e m. Determinar a equação do seu movimento, sendo eles sujeitos apenas à atração gravitacional mútua.

### _Solução_

Usando vetores no espaço em três dimensões, sendo a posição do corpo A o vetor $\mathbf {x} _{A}$ , a posição do corpo B o vetor $\mathbf {x} _{B}$, as acelerações os vetores $\mathbf {a} _{A}$ e $\mathbf {a} _{B}$, a força exercida sobre A o vetor $\mathbf {F} _{A}$ e a força exercida sobre B o vetor $\mathbf {F} _{B}$, temos:

\begin{equation}
{F} _{A}=M\ \mathbf {a} _{A} \\
{F} _{B}=m\ \mathbf {a} _{B} \\
\end{equation}

Pela lei da gravitação universal, escrita em forma vetorial, temos:

\begin{equation}
 {F} _{A}=-G\ m\ M\ {\frac {\mathbf {x} _{A}-\mathbf {x} _{B}}{\left|\mathbf {x} _{A}-\mathbf {x} _{B}\right|^{3}}} \\ \mathbf {F} _{B}=-G\ m\ M\ {\frac {\mathbf {x} _{B}-\mathbf {x} _{A}}{\left|\mathbf {x} _{B}-\mathbf {x} _{A}\right|^{3}}} 
\end{equation}

Assim, temos que a aceleração relativa do corpo B em relação ao corpo A, $-\mathbf {a} _{A}$ $-\mathbf {a} _{A}$, pode ser escrita como:

\begin{equation}
{\displaystyle \mathbf {a} _{B}-\mathbf {a} _{A}=-G\ (M+m)\ {\frac {\mathbf {x} _{B}-\mathbf {x} _{A}}{\left|\mathbf {x} _{B}-\mathbf {x} _{A}\right|^{3}}}\,}
\end{equation}
Ou seja, reduzimos o problema dos dois corpos ao estudo do movimento de um corpo de massa desprezível submetido à atração gravitacional de um corpo de massa M + m.

#### Propriedades da solução
Como o sistema não é influenciado por forças externas, seu centro de massa possui aceleração nula:

\begin{equation}
{\displaystyle {\frac {d^{2}}{dt^{2}}}\left({\frac {M\mathbf {x} _{A}+m\mathbf {x} _{B}}{m+M}}\right)={\frac {M\mathbf {a} _{A}+m\mathbf {a} _{B}}{m+M}}={\frac {\mathbf {F} _{A}+\mathbf {F} _{B}}{m+M}}=0\,}
\end{equation}

- Como a força gravitacional é uma força central, o momento angular é constante.
- Como a força gravitacional é conservativa, a energia do sistema é constante.

A seguir, vamos ver o problema de dois corpos compilado em linguagem Python em duas dimensões e em seguida em três dimensões. 

A biblioteca utilizada na solução do  Problema de Dois Corpos em 3D é Scipy que soluciona a EDO pelo método chamado 'lsoda', disponível em: <https://docs.scipy.org/doc/scipy/reference/generated/scipy.integrate.LSODA.html#scipy.integrate.LSODA> Consultado em: 10/10/2021

#### Problema de Dois Corpos em 2D

```python
#aqui importamos as bibliotecas que serão utilizadas
import numpy as np
import matplotlib.pyplot as plt
```

```python

dia = 60*60*24

# Constantes
G = 6.67408e-11
UA = 1.496e11

class CelBody(object):
    # Constante da Gravitação Universal
    def __init__(self, id, nome, x0, v0, massa, color, lw):
        # Nome do corpo (string)
        self.id = id
        self.nome = nome
        # Massa do corpo (kg)
        self.Mass = massa
        # Posição inicial do corpo(UA)
        self.x0 = np.asarray(x0, dtype=float)
        # Posição (UA). Define o valor inicial.
        self.x = self.x0.copy()
        # Velocidade inicial do corpo (UA/s)
        self.v0 = np.asarray(v0, dtype=float)
        # Velocidade (UA/s). Define o valor inicial
        self.velocidade = self.v0.copy()
        self.aceleracao = np.zeros([3], dtype=float)
        self.color = color
        self.lw = lw

# Todos os corpos celestes

t = 0
dt = 0.1*dia

Corpos = [
    CelBody(0, 'Sol', [0, 0, 0], [0, 0, 0], 1.989e30, 'yellow', 10),
    CelBody(1, 'Terra', [-1*UA, 0, 0], [0, 29783, 0], 5.9742e24, 'blue', 3),
    ]

paths = [ [ b.x[:2].copy() ] for b in Corpos]

#loop ao longo de vinte anos astronômicos
v = 0
while t < 20*365.242*dia:
    # computa forças/acelerações
    for corpo in Corpos:
        corpo.aceleracao *= 0
        for other in Corpos:
            if (corpo == other): continue # pula para o próximo loop
            rx = corpo.x - other.x
            r3 = sum(rx**2)**1.5
            corpo.aceleracao += -G*other.Mass*rx/r3

    for n, planeta in enumerate(Corpos):
        #usa o método de Euler (sympletic) para melhor conservação das constantes de movimento
        planeta.velocidade += planeta.aceleracao*dt
        planeta.x += planeta.velocidade*dt
        paths[n].append( planeta.x[:2].copy() )
        if t > v:
            v += 30.5*dia
        t += dt

plt.figure(figsize=(10,10))
for n, planeta in enumerate(Corpos): 
    px, py=np.array(paths[n]).T; 
    plt.plot(px, py, color=planeta.color, lw=planeta.lw)
plt.show()
```

#### Problema de Dois Corpos em 3D

```python
#aqui importamos as bibliotecas que serão utilizadas
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import ode
from mpl_toolkits.mplot3d import Axes3D
```

```python
def plot(rs):
    fig = plt.figure()
    ax = fig.add_subplot(111,projection='3d')
    #Resolvendo a matriz de entrada 
    x1 = rs[:,0]
    y1 = rs[:,1]
    z1 = rs[:,2]
    x2 = rs[:,3]
    y2 = rs[:,4]
    z2 = rs[:,5]
    xG = rs[:,6]
    yG = rs[:,7]
    zG = rs[:,8]
    
    ax.set_xlabel('X(km)')
    ax.set_ylabel('Y(km)')
    ax.set_zlabel('Z(km)')

    #Movimento relativo a um referencial inercial 
    
    ax.plot(x1,y1,z1,'g',label="Trajetória de m1")
    ax.plot(xG,yG,zG,'r',label="Trajetória do CM")
    ax.plot(x2,y2,z2,'b',label="Trajetória de m2")
    ax.plot([rs[0,0]],[rs[0,1]],[rs[0,2]],'go',label='Posição inicial de m1')
    ax.plot([rs[0,6]],[rs[0,7]],[rs[0,8]],'ro',label='Posição inicial do CM')
    ax.plot([rs[0,3]],[rs[0,4]],[rs[0,5]],'bo',label='Posição inicial de m2')
    #Plotando os vetores
    _x,_y,_z = [[0,0,0],[0,0,0],[0,0,0]]
    _u,_v,_w = [[rs[len(rs[:,0])-1,0],0,0],[0,rs[len(rs[:,1])-1,1],0],[0,0,rs[len(rs[:,2])-1,2]]]
    ax.quiver(_x,_y,_z,_u,_v,_w,color='k')
    ax.set_title('Movimento de dois corpos, visto de um referencial inercial ')
    plt.legend()
    
    #Movimento relativo ao centro de massa
    
    fig = plt.figure()
    ax = fig.add_subplot(111,projection='3d')
    ax.plot(x1-xG,y1-yG,z1-zG,'g',label='Trajetória de m1')
    ax.plot(x2-xG,y2-yG,z2-zG,'b',label='Trajetória de m2')
    
    ax.plot([(x1-xG)[0]],[(y1-yG)[0]],[(z1-zG)[0]],'go',label='Posição inicial de m1')
    ax.plot([0],[0],[0],'ro', label='CM')
    ax.plot([(x2-xG)[0]],[(y2-yG)[0]],[(z2-zG)[0]],'bo',label='Posição inicial de m2')
    #Plotando os vetores
    _x,_y,_z = [[0,0,0],[0,0,0],[0,0,0]]
    _u,_v,_w = [[abs((x1-xG)[len(x1-xG)-1]),0,0],[0,abs((y1-yG)[len(y1-yG)-1]),0],[0,0,abs((z1-zG)[len(z1-zG)-1])]]
    ax.quiver(_x,_y,_z,_u,_v,_w,color='k')
    plt.legend()
    ax.set_title("Movimento em relação ao centro de massa CM")
    ax.set_xlabel('X(km)')
    ax.set_ylabel('Y(km)')
    ax.set_zlabel('Z(km)')
    
    #Movimento relativo a massa m1
    
    fig = plt.figure()
    ax = fig.add_subplot(111,projection='3d')
    ax.plot(x2-x1,y2-y1,z2-z1,'b',label='Trajetória de m2')
    ax.plot(xG-x1,yG-y1,zG-z1,'r',label='Trajetória do CM')
    
    ax.plot([(x2-x1)[0]],[(y2-y1)[0]],[(z2-z1)[0]],'bo',label='Posição inicial de m2')
    ax.plot([0],[0],[0],'go', label='m1')
    ax.plot([(xG-x1)[0]],[(yG-y1)[0]],[(zG-z1)[0]],'ro',label='Posição inicial do CM')
    #Vectors plot
    _x,_y,_z = [[0,0,0],[0,0,0],[0,0,0]]
    _u,_v,_w = [[abs((x2-x1)[len(x2-x1)-1]),0,0],[0,abs((y2-y1)[len(y2-y1)-1]),0],[0,0,abs((z2-z1)[len(z2-z1)-1])]]
    ax.quiver(_x,_y,_z,_u,_v,_w,color='k')
    
    ax.set_title("Movimento em relação massa m1")
    ax.set_xlabel('X(km)')
    ax.set_ylabel('Y(km)')
    ax.set_zlabel('Z(km)')
    
    plt.legend() #aqui plotamos a legenda do gráfico
    plt.show() #aqui plotamos o gráfico
```

```python
#A função que retorna a velocidade e a aceleração
def diff(t,y):
    #Resolvendo o estado
    r1 = np.array([y[0],y[1],y[2]])
    r2 = np.array([y[3],y[4],y[5]])
    
    v1 = [y[6],y[7],y[8]]
    v2 = [y[9],y[10],y[11]]
    
    #Norma de r2-r1
    r = np.linalg.norm(r2-r1)
    
    A1 = (G*m2*(r2-r1))/(r**3)
    A2 = (G*m1*(r1-r2))/(r**3)
    
    estado_final = v1 + v2 + list(A1) + list(A2)
    return estado_final
```

```python
G = 6.67259e-20
#Massas dos corpos
m1 = 1e26
m2 = 1e26

#Tempo inicial
t0 = 0

#Posição inicial das massas
r1_0 = [0,0,0]
r2_0 = [3000,0,0]

#Velocidades iniciais das massas
v1_0 = [10,20,30]
v2_0 = [0,40,0]

#Vetor do estado inicial
y0 = r1_0 + r2_0 + v1_0 + v2_0
```

Método Scipy de solução da EDO pelo método chamado 'lsoda'
O objeto do solucionador de equação diferencial

```python
solu = ode(diff)
solu.set_integrator('lsoda')
solu.set_initial_value(y0,0)
```

```python
#intervalo de tempo
t_int = 1000
dt = 1.0
    
n_passos = int(np.ceil((t_int-t0)/dt))
    
#Inicializando arranjos
ys = np.zeros((n_passos,12))
ts = np.zeros((n_passos,1))
ys[0] = np.array(y0)
passo = 1
```

```python
#Propagação da Solução da Equação Diferencial
while solu.successful() and passo<n_passos:
    solu.integrate(solu.t + dt)
    ts[passo] = solu.t
    ys[passo] = solu.y
    passo+=1

x1 = ys[:,0]
y1 = ys[:,1]
z1 = ys[:,2]
x2 = ys[:,3]
y2 = ys[:,4]
z2 = ys[:,5]   

xG = (m1*x1 + m2*x2)/(m1+m2)
yG = (m1*y1 + m2*y2)/(m1+m2)
zG = (m1*z1 + m2*z2)/(m1+m2)


rs = ys[:,:6]


rs = np.column_stack((rs,xG,yG,zG))
plot(rs)
```
