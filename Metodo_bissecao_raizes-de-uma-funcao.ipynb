## Exemplo de método da bisseção para encontrar raízes de uma função


Determine as raízes da função abaixo:

$e^x - sen(x) - 2x^2$


##### _Resposta 1_

```python
#aqui importamos as bibliotecas que serão utilizadas
import math 
import numpy as np
import matplotlib.pyplot as plt 
```

```python
# permite que se mostre os gráficos nesse ambiente
%matplotlib inline
```

```python
# aqui fazemos o gráfico da função


# Definindo a função
def f(x):
    return np.exp(x) - np.sin(x) - 2*x**2

x0 = -2      # Valor inicial de x
xf = 3       # Valor final de x
dx = 0.01     # Passo da discretização dx = (xf - x0)/(np-1)
NP =101
#x = np.arange(x0, xf, dx)      # cria um vetor de dados x-data
x = np.linspace(x0,xf,NP)
y = f(x)                        # calcula o vetor de dados y-data
y1 = np.exp(x)
y2 = np.sin(x)
y3 = 2*x**2

largura = 20    # Largura da figura
altura  = 10     # Altura da figura
plt.figure(figsize =(largura, altura))     # Define o tamanho do gráfico
plt.grid()                                 # Habilita a grade 
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Função transcendental f(x)=exp(x) - sen(x) - 2x^2')
plt.plot(x, y, label="exp(x)- sen(x) - 2x^2")    # Faz o gráfico
plt.plot(x, y1, label="exp(x)")      # Faz o gráfico
plt.plot(x, y2, label="sen(x)")          # Faz o gráfico
plt.plot(x, y3, label="2x^2")       #Faz o gráfico
plt.legend(loc="upper left")         # Habilita as legendas
plt.show()                           #plota o gráfico
```

```python
# após observar o comportamento do gráfico, podemos ter noção dos limites das raízes e usar o Método da Bisseção

def Bissecao(f, a, b, prec=1.0e-8, nmax=100 ):
    """
    Método da bisseção para encontrar as raízes de uma função f
    que possui pelo menos uma raíz no intervalo [a,b].
    """
    # Verificar se no intervalo dado há uma raiz
    fa = f(a)
    fb = f(b)
    if fa*fb > 0.0 : 
        return print("Não há garantias de haver uma raiz no intervalo")
    else: 
        m =(a+b)/2.0
        fm = f(m)
        d = b - a
        cont = 0
        while (cont <= nmax) :
            cont = cont + 1
            if fa * fm < 0:
                b = m
                fb = fm
            else:
                a = m
                fa =fm                
            m  = (a + b)/2
            fm = f(m)
            d   = b - a
            if ( (abs(fm) < prec) or ( abs(d) < prec ) ):  
                return m, cont
        else:
            print(f'Não foi achada uma raiz após {cont} iterações, com a precisão desejada de {prec}.\nO valor de f({m})={fm}.')
            return m, cont
```

```python
# Aqui definimos a função f
def f(x):
    return np.exp(x) - np.sin(x) - 2*x**2
```

```python
# Obtendo a raiz da função f(x) = exp(x)-sin(x) -2*x^2 no intervalo de [-2,0]
xe = -2.0            # Ponto a esquerda da raiz da função f(x)
xd = 0.0            # Ponto a direita da raiz da função f(x)
Erro = 1.0e-10      # Tolerância ou erro máximo admitido
ni  =  100          # Número máximo de iterações permitidas
raiz, numero_iteracoes = Bissecao(f, xe, xd, Erro, ni)
print(f'A raiz é {raiz} e  foi encontrada com {numero_iteracoes} iterações' )
```
